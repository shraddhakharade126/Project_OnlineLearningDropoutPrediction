```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import dataset 

df = pd.read_csv("newdataset.csv")
df
```

```python
df = df.drop(columns =['Unnamed: 0'], errors ='ignore')
df
```

```python
sns.countplot(x='dropout', data=df)
plt.title("Dropout Distribution (0 = No, 1 = Yes)")
plt.show()
```

```python
df.columns
```

```python

plt.figure(figsize=(6,6))
sns.histplot(df['login_frequency_per_week'], kde=True ,bins =15)
plt.title(" login_frequency_per_week ")
plt.xlabel("frequency")
plt.show()
```

```python
sns.pairplot(df.sample(200), hue='dropout')
plt.show()
```

```python
sns.boxplot(x=df['emails_sent_per_day'])
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot= True , cmap='coolwarm' ,fmt= " .2f")
plt.title("Feature Correlation HeatMap ")
plt.show()
```
